```python
verts = [list(v) for v in vertdata]
verts = np.asarray(verts)
```

### Look at the bunny point cloud from different viewpoints

```python
fig = plt.figure(figsize=(20,20))
ds = 2 # downsampling ratio
elems = list(range(0, 180, 30))
azims = list(range(0, 360, 60))
i = 1
for row, elem in enumerate(elems):
    for col, azim in enumerate(azims):
        ax = fig.add_subplot(len(azims), len(elems), i, projection='3d')
        ax.scatter(verts[::ds,0], verts[::ds,1], verts[::ds,2], s=1)
        ax.view_init(elem, azim)
        plt.axis('off')
        plt.title('['+str(elem)+', '+str(azim)+']')
        i = i + 1
plt.tight_layout()
```
